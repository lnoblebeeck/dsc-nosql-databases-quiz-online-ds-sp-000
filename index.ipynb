## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

The difference between NoSQL and SQL databases is that later is used to store relational data and the latter is used for Non-Relational data based on dynamic schema. 
No one is better then the other, rather it depends more on the type of data we are looking to store/access. While SQL is arguably more powerful computationally, NoSQL is easier to scale due to its ability to store data on multiple servers.

NoSQL data might be better suited for keeping track of data pertaining to a large retail coorperation with multiple sale points. ## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

MongoDB works as a document based stores, that is to say, it stores data as unique documents with the accompanying metadata being stored within. The information within each document is stored within Key value pairs that make accessing and exploring the data more straightforward. The versatility of this storage alternative is that each document can have its own Key value pairs (main difference with relational databases). Furthermore, unlike with relational databases where the schema must be prepared before, document based storage key values pairs (the equivaleent to columns) can be created on the fly.






## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Fault tolerance refers to the ability of a system to function after one of its component fails. This is crucial component of resilient distributed datasets. A RDDs is a collection of an arbitrary number of server clusters that share a full dataset distributed between them, by sharing information between then, they become fault reesistent to an extent. By keeping random "copies" of different elements of the dataset distributed randomly among the cluster of servers, we can ensure taht should one server malfuction, the data that is stored there can still be access, through its "copies".
